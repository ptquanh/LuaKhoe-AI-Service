# Rice Disease Detection Training & ONNX Export

This notebook is designed for training a MobileNetV3 model on Google Colab and exporting it to ONNX format for the **Lúa Khỏe AI** backend.

```python
!pip install torch torchvision onnx onnxruntime
import torch
import torchvision
from torchvision import datasets, models, transforms
import torch.nn as nn
```

## 1. Load Pretrained Model

```python
model = models.mobilenet_v3_small(pretrained=True)
# Change head for 5 classes (Healthy, BLB, Brown Spot, Blast, Smut)
num_ftrs = model.classifier[3].in_features
model.classifier[3] = nn.Linear(num_ftrs, 5)
model.eval()
```

## 2. Export to ONNX

```python
dummy_input = torch.randn(1, 3, 224, 224)
torch.onnx.export(model, dummy_input, "v1_cnn_mvp.onnx", verbose=True)
```
